```python
import pandas as pd
import geopandas as gpd
```

```python
df_restaurants_review = pd.read_csv('../../data/raw/restaurant/restaurants.csv')
gdf_restaurants = gpd.GeoDataFrame(df_restaurants_review, geometry=gpd.points_from_xy(df_restaurants_review.lng, df_restaurants_review.lat)).set_crs(epsg=4326)
```

```python
gdf_restaurants.head()
```

```python
# filtering buildings with the Manhattan boundary
gdf_manhattan = gpd.read_file('../../data/raw/boundary/Borough Boundaries.geojson')
gdf_manhattan = gdf_manhattan.loc[gdf_manhattan.loc[:,'boro_name'] == 'Manhattan']




gdf_restaurants_manhattan = gpd.sjoin(gdf_restaurants, gdf_manhattan, how='inner', op='intersects')
gdf_restaurants_manhattan = gdf_restaurants_manhattan.drop(['index_right', 'boro_code', 'boro_name', 'shape_area','shape_leng'], axis=1)
```

```python
gdf_restaurants_manhattan.head()
```

```python
gdf_buildings = gpd.read_file('../../data/processed/building/building_manhattan_commercial_vt.geojson')
```

```python
gdf_buildings.head()
```

```python
gdf_restaurants_with_bin = gdf_restaurants_manhattan.to_crs(2263).sjoin_nearest(gdf_buildings.loc[:,['bin','geometry']], how='left', distance_col='distance')
```

```python
gdf_restaurants_with_bin.drop(['geometry','index_right'], axis=1).to_csv('../../data/processed/restaurant_review.csv', index=False)
```

```python
gdf_restaurants_with_bin.loc[gdf_restaurants_with_bin.loc[:,'distance']<10].drop(['geometry','index_right'], axis=1).to_csv('../../data/processed/restaurant_review_240405.csv', index=False)
```
